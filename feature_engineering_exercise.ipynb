# Prediction of sales

### Problem Statement
[The dataset](https://drive.google.com/file/d/1B07fvYosBNdIwlZxSmxDfeAf9KaygX89/view?usp=sharing) represents sales data for 1559 products across 10 stores in different cities. Also, certain attributes of each product and store are available. The aim is to build a predictive model and find out the sales of each product at a particular store.

|Variable|Description|
|: ------------- |:-------------|
|Item_Identifier|Unique product ID|
|Item_Weight|Weight of product|
|Item_Fat_Content|Whether the product is low fat or not|
|Item_Visibility|The % of total display area of all products in a store allocated to the particular product|
|Item_Type|The category to which the product belongs|
|Item_MRP|Maximum Retail Price (list price) of the product|
|Outlet_Identifier|Unique store ID|
|Outlet_Establishment_Year|The year in which store was established|
|Outlet_Size|The size of the store in terms of ground area covered|
|Outlet_Location_Type|The type of city in which the store is located|
|Outlet_Type|Whether the outlet is just a grocery store or some sort of supermarket|
|Item_Outlet_Sales|Sales of the product in the particulat store. This is the outcome variable to be predicted.|

Please note that the data may have missing values as some stores might not report all the data due to technical glitches. Hence, it will be required to treat them accordingly.



### Explore the problem in following stages:

1. Hypothesis Generation – understanding the problem better by brainstorming possible factors that can impact the outcome
2. Data Exploration – looking at categorical and continuous feature summaries and making inferences about the data.
3. Data Cleaning – imputing missing values in the data and checking for outliers
4. Feature Engineering – modifying existing variables and creating new ones for analysis
5. Model Building – making predictive models on the data

```python
import pandas as pd

#Read files:
data = pd.read_csv("regression_exercise.csv", delimiter=',')
data
```

```python
# Filling missing values for Item_Weight variable
for type in set(data.Item_Type):
    data.loc[data['Item_Type'] == type, 'Item_Weight'] = data.loc[data['Item_Type'] == type, 'Item_Weight'].fillna(data['Item_Weight'][data['Item_Type'] == type].mean())
```

```python
# Filling missing values for Outlet_Size variable
data.loc[:,'Outlet_Size'] = data['Outlet_Size'].fillna('Unknown')
```

```python
data
```

## 4. Feature Engineering

1. Resolving the issues in the data to make it ready for the analysis.
2. Create some new variables using the existing ones.





### Create a broad category of Type of Item

Item_Type variable has many categories which might prove to be very useful in analysis. Look at the Item_Identifier, i.e. the unique ID of each item, it starts with either FD, DR or NC. If you see the categories, these look like being Food, Drinks and Non-Consumables. Use the Item_Identifier variable to create a new column

```python
Item_Category = []
for item in data['Item_Identifier']:
    Item_Category.append(item[0:2])

data['Item_Category'] = Item_Category
```

```python
data.loc[data['Item_Category'] == 'FD', 'Item_Category'] = 'Food'
data.loc[data['Item_Category'] == 'DR', 'Item_Category'] = 'Drinks'
data.loc[data['Item_Category'] == 'NC', 'Item_Category'] = 'Non-Consumables'
```

### Determine the years of operation of a store

Make a new column depicting the years of operation of a store (i.e. how long the store exists). 

```python
data['Outlet_Operation_Years'] = 2020 - data['Outlet_Establishment_Year']
```

### Modify categories of Item_Fat_Content

There are difference in representation in categories of Item_Fat_Content variable. This should be corrected.

```python
data.loc[data['Item_Fat_Content'].isin(('LF', 'Low Fat', 'low fat')), 'Item_Fat_Content'] = 'Low Fat'
data.loc[data['Item_Fat_Content'].isin(('Regular', 'reg')), 'Item_Fat_Content'] = 'Regular'
```

There are some non-consumables as well and a fat-content should not be specified for them. Create a separate category for such kind of observations.

```python
# Setting item fat content variable for non-consumables to NA (Not Applicable)
data.loc[data['Item_Category'] == 'Non-Consumables', 'Item_Fat_Content'] = 'NA'
```

### Numerical and One-Hot Encoding of Categorical variables

Since scikit-learn algorithms accept only numerical variables, we need to convert all categorical variables into numeric types. 

- if the variable is Ordinal we can simply map its values into numbers
- if the variable is Nominal (we cannot sort the values) we need to One-Hot Encode them --> create dummy variables

```python
# ordinal variables
data = data.replace({'Item_Fat_Content' : {'NA' : 0, 'Low Fat' : 1, 'Regular' : 2},
                     'Outlet_Size' : {'Unknown' : 0, 'Small' : 1, 'Medium' : 2, 'High' : 3}}
                   )
```

```python
# Item category dummy variables
data.loc[data['Item_Category'] == 'Drinks', 'Drinks'] = 1
data.loc[data['Item_Category'] != 'Drinks', 'Drinks'] = 0

data.loc[data['Item_Category'] == 'Food', 'Food'] = 1
data.loc[data['Item_Category'] != 'Food', 'Food'] = 0

#outlet identifier dummy variables
data.loc[data['Outlet_Identifier'] == 'OUT010', 'OUT010'] = 1
data.loc[data['Outlet_Identifier'] != 'OUT010', 'OUT010'] = 0

data.loc[data['Outlet_Identifier'] == 'OUT013', 'OUT013'] = 1
data.loc[data['Outlet_Identifier'] != 'OUT013', 'OUT013'] = 0

data.loc[data['Outlet_Identifier'] == 'OUT017', 'OUT017'] = 1
data.loc[data['Outlet_Identifier'] != 'OUT017', 'OUT017'] = 0

data.loc[data['Outlet_Identifier'] == 'OUT018', 'OUT018'] = 1
data.loc[data['Outlet_Identifier'] != 'OUT018', 'OUT018'] = 0

data.loc[data['Outlet_Identifier'] == 'OUT019', 'OUT019'] = 1
data.loc[data['Outlet_Identifier'] != 'OUT019', 'OUT019'] = 0

data.loc[data['Outlet_Identifier'] == 'OUT027', 'OUT027'] = 1
data.loc[data['Outlet_Identifier'] != 'OUT027', 'OUT027'] = 0

data.loc[data['Outlet_Identifier'] == 'OUT035', 'OUT035'] = 1
data.loc[data['Outlet_Identifier'] != 'OUT035', 'OUT035'] = 0

data.loc[data['Outlet_Identifier'] == 'OUT045', 'OUT045'] = 1
data.loc[data['Outlet_Identifier'] != 'OUT045', 'OUT045'] = 0

data.loc[data['Outlet_Identifier'] == 'OUT046', 'OUT046'] = 1
data.loc[data['Outlet_Identifier'] != 'OUT046', 'OUT046'] = 0

#outlet location type dummy variables
data.loc[data['Outlet_Location_Type'] == 'Tier 1', 'Tier_1'] = 1
data.loc[data['Outlet_Location_Type'] != 'Tier 1', 'Tier_1'] = 0

data.loc[data['Outlet_Location_Type'] == 'Tier 2', 'Tier_2'] = 1
data.loc[data['Outlet_Location_Type'] != 'Tier 2', 'Tier_2'] = 0

#outlet type
data.loc[data['Outlet_Type'] == 'Grocery Store', 'Grocery_Store'] = 1
data.loc[data['Outlet_Type'] != 'Grocery Store', 'Grocery_Store'] = 0

data.loc[data['Outlet_Type'] == 'Supermarket Type1', 'Supermarket_Type1'] = 1
data.loc[data['Outlet_Type'] != 'Supermarket Type1', 'Supermarket_Type1'] = 0

data.loc[data['Outlet_Type'] == 'Supermarket Type2', 'Supermarket_Type2'] = 1
data.loc[data['Outlet_Type'] != 'Supermarket Type2', 'Supermarket_Type2'] = 0

data
```
